```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

from tqdm.notebook import tqdm
#индикатор обновления
tqdm.pandas()

import warnings
warnings.filterwarnings("ignore")
```

```python
path = 'wineQualityReds'
winered = pd.read_csv('wineQualityReds.csv', index_col='App')
winered['wine'] = 'Red'
winered
```

```python
path = 'wineQualityWhites'
winewhite = pd.read_csv('wineQualityWhites.csv', index_col='App')
winewhite['wine'] = 'white'
winewhite
```

```python
raw_data = pd.concat([winered, winewhite], ignore_index=True)
raw_data = raw_data.rename_axis('App')
raw_data
```

```python
raw_data.isnull().sum(axis = 0)
```

```python
raw_data.info()
```

```python
raw_data.describe()
```

```python
sns.catplot(data=raw_data, x='fixed.acidity', y='volatile.acidity', height=10)
```

```python
sns.pairplot(data=raw_data, hue='quality')
```

```python
numeric_col = ['chlorides', 'free.sulfur.dioxide', 'total.sulfur.dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
corr_matrix = raw_data.loc[:,numeric_col].corr()
print(corr_matrix)
#визуализация
sns.heatmap(corr_matrix, annot=True)
#Создание корреляционной матрицы в Python 	'fixed.acidity', 'volatile.acidity', 'citric.acid', 'residual.sugar', 						
```
